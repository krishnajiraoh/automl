```python
churn_df = pd.read_csv("data/churn_data.csv")
churn_df.head()
```

```python
X = churn_df.iloc[: , 1:-1]
y = churn_df.iloc[: , -1]

X_train, X_test, y_train, y_test = \
    sklearn.model_selection.train_test_split(X, y, random_state=1)
```
